## NBA Roster Tool
Note, there will be an interactive website where you can play around with the tool's outputs (ongoing). Below is meant to explain what is happening in more detail.

The tool is designed to take any combinations of players, then (1) summarize the team (positions, minutes played per games, an individual rating "RPMScore" and show the points per game; then (2) look at the roster's strength and weaknesses using various unsupervised learning methods to cluster the data and find which combinations of features are most import in explaining the data (eg, PCA, K-means, etc.); and finally (3) use an aggregation of several supervised machine learning models to predict all given players' expected performance if they were to PLAY ON THAT TEAM.

The "Individual Rating" or "RPM Score" are naive scores that JUST use a player's individual statistics and try to project the player's performance from there. What this tool does is add thousands of potential features at the individual and team level (including how a player's individual metrics INTERACT with the team's attributes), and then find those players that benefit the MOST from playing on your given roster (i.e. the input). The tool is really trying only to maximize the "fit" and NOT the overall rating. The idea is the overall rating ("Individual Rating") gives you a good idea of how good the players is, and then you can use this tool to find the players that see the biggest "Team Gain" factor, which is ADDITIONAL performance due to the player-team combination.

In practice, it becomes clear that the tool (unsuprisingly) produces players that are quite similar with each other. This is because it finds your team's needs, and produces the top 10 players that best benefit from your team's strengths (and make up for the weaknesses). I'll go over a few examples, starting with the Washington Wizards roster at the end of 2022-23 (it has changed a lot since then) and then giving a custom team example where I take the Philadelphia 76ers current projected starting lineup.

Of course, there are many things happening under the hood, and many combinations taken into account, so it won't always be easy to identify why a given player would do well. But rest assured the model is quite effective at predicting which players might complement the given roster well! 

### Some rules of thumb: 
(1) Defense and effort (eg, rebounding, hustle players) scale better than offense. Having only one defensive player is bad. Two is better, and three or four is also better. Very few diminshing returns. Meanwhile, there are sharp diminishing return to offense-first players (the ideal is 2 or 3. 4 good offensive players that are not good at defense will be detrimental. (2) Highly talented, but inexperienced players, benefit A LOT from an experienced roster. (3) Steals are quite important! (4) Having two scorers/play-makers (rather than just one) helps a team a lot. Three not so much. (5) A very good defensive center is very useful to make up for defensive efficiencies of others. (6) It turns out that being a "good enough" shooter from long range (3-point shots) is not much worse than being very good. It seems the player needs to be good enough to force the defense to pay attention to him, and that is most of the value of spacing/shooting. (7) Roster balance is generally good. You don't want too many similar players. An exception is "effort guys", as mentioned in point (1).
